```python
mix_match_elements = ['apple', 280, 'mango', 200, 'kiwi', 120]

total_amount = 0
count = 1

while count < len(mix_match_elements):
    amount = mix_match_elements[count]
    
    total_amount = total_amount + amount
    
    count = count + 2

print('I am outside loop')
print('total amount is ', total_amount)
```

```python
# fact data

vitamin_c_rich_foods = ['orange', 'kiwi', 'lemon', 'pineapple']

vitamin_a_rich_foods = ['carrot', 'mango', 'papaya', 'apricot']

vitamin_b12 = ['fish', 'prawns']

prtein_foods = ['chicken', 'mutton', 'fish', 'prawns', 'lentils', 'rajma', 'soya']
```

```python
my_food_basket = ['orange', 'apple', 'pineapple', 'mango', 'kiwi', 'chicken']

for food in my_food_basket:
    if food in vitamin_c_rich_foods:
        print('Vitamin C')
    if food in vitamin_a_rich_foods:
        print('Vitamin A')
    if food in vitamin_b12:
        print('Vitamin B12')
    if food in prtein_foods:
        print('Protein')
```

```python
# Break and continue statements

# find the first even number in this list?

numbers = [1, 3, 5, 8, 11, 14]

for num in numbers:
    print(f'Looking at num {num}')
    if num % 2 == 0:
        print(f"First Even number found: {num}")
        break
```

```python
# continue


numbers = [10, -3, 25, -8, 100, -500, 23, 45]


# Aim : Skip - numbers

for num in numbers:
    if num < 0:
        continue
    print(f'Looking at num {num}')
```

```python
marks = [89, 78]

marks_acquired = sum(marks)

out_of = 200

per = (marks_acquired/out_of)*100

print(per)
```

```python
def calculate_percentage(marks_acquired, out_of):
    per = (marks_acquired/out_of)*100
    return per
```

```python
DRY Principle:
    Do not Repeat Yourself
    The error happening in your code reduces drastically.
```

```python
students = {'Tanveer': [89, 78], 'Pradeep': [66, 99], 'Gaurav': [88, 89], 
            'Hitakshi': [99, 98], 'Asraful': [75, 79]}


for stu, marks in students.items():
    print(stu)
    percentage = calculate_percentage(sum(marks), 200)
    print(percentage)
```
