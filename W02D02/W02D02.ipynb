```python
# Lambda Functions

# single line functions for very small usecases.
# They work like a shortcut for doing small functional work.
# They are also called as nameless functions.
# The reason we don't usually have name for them is because they are mostly single time use.
# lambda functions are used mostly in inbuilt functions like sorted, map, filter, etc
```

```python
# sorted() inbuilt function

x = ['pradeep', 'tanveer', 'dsrose']

result = sorted(x)

print(result)

y = [85, 92, 78]

result2 = sorted(y, reverse=True)

print(result2)
```

```python
students = [("pradeep", 85), ("tanveer", 92), ("dsrose", 78)]

# alphabetically on name, order will be dsrose, pradeep, tanveer

# but marks wise tanveer, pradeep, dsrose
```

```python
def give_correct_item_for_sort(item_pair):
    return item_pair[1]
```

```python
students = [("pradeep", 85), ("tanveer", 92), ("dsrose", 78)]


result = sorted(students, reverse=True, key=give_correct_item_for_sort)
```

```python
print(result)
```

```python
# Syntax:-

lambda input_argument : output_single_line_code(which acts as an return value)
```

```python
students = [("pradeep", 85), ("tanveer", 92), ("dsrose", 78)]


result = sorted(students, reverse=True, key=lambda item_pair : item_pair[1])
```

```python
# Scope and Namespaces

# Local vs Global Scope
```

```python
The life of global variable is till the program or python code ends.
```

```python
balance = 1000 # balance is a global variable

# if you want to access a global variable inside function you should use the keyword global to define it first
# inside the function


def deposit(amount):
    global balance
    balance += amount
    print(f"New balance: {balance}")

deposit(200)
```

```python
deposit(200)
```

```python
The life of local variable is only till the function call ends.
So if you are calling a function 10 times, you are actually 
creating 10 different local variable with same name and destryong them once the
function ends.
```

```python
balance = 1000 # balance is a global variable

# if you want to access a global variable inside function you should use the keyword global to define it first
# inside the function


def deposit(amount):
    balance = 0  # local balance variable
    balance += amount
    print(f"New balance: {balance}")

deposit(200)
```

```python
deposit(400)
```

```python
# E-commerce 

products = [
    {'name': 'Laptop', 'price': 75000, 'category': 'Electronics'},
    {'name': 'Phone', 'price': 25000, 'category': 'Electronics'},
    {'name': 'Shirt', 'price': 1500, 'category': 'Clothing'},
    {'name': 'Watch', 'price': 8000, 'category': 'Accessories'},
    {'name': 'Shoes', 'price': 3000, 'category': 'Clothing'},
    {'name': 'Earphone', 'price': 1000, 'category': 'Electronics'}
]

my_products = [p['name'] for p in products]

print(my_products)
```

```python
# list comprehension for finding electronics < 50000

affordable_electronics = [
    p['name'] for p in products if p['category'] == 'Electronics' and p['price'] < 50000
]

print(affordable_electronics)
```

```python
# data population

facebook = []

instagram = {}

for no in range(1, 10**7):
    name = f"PersonName{no}"
    facebook.append(name)
    instagram[name] = True
```

```python
def search_facebook(name):
    found = False
    for any_name in facebook:
        if any_name == name:
            found = True
            break
    return found
```

```python
def search_instagram(name):
    result = instagram.get(name)
    if result is True:
        return True
    else:
        return False
```

```python
# I met somebody and I want to search them on social media.

name = 'PersonName9999999'
```

```python
import time
```

```python
start_time = time.perf_counter()

search_result = search_facebook(name)

end_time = time.perf_counter()

execution_time1 = end_time - start_time
print(f"Execution time: {execution_time1:.6f} seconds")

print(search_result)
```

```python
start_time = time.perf_counter()

search_result = search_instagram(name)

end_time = time.perf_counter()

execution_time2 = end_time - start_time
print(f"Execution time: {execution_time2:.6f} seconds")

print(search_result)
```

```python
multiplier = execution_time1/execution_time2

print(f'The dictionary in python is {multiplier} times fast in this case!')
```
