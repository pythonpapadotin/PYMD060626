```python
class NewStudentRegistration:
    def __init__(self, student_first_name, student_last_name, parent_first_name, parent_last_name,
                 email):
        self.student_first_name = student_first_name
        self.student_last_name = student_last_name
        self.parent_first_name = parent_first_name
        self.parent_last_name = parent_last_name
        self.email = email

    def send_email(self, heading, content):
        print(f'Email to {self.student_first_name} sending.....{heading}....{content[0:5]}....')
```

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print(f'self value is {self}')
        return f"{self.name} says woof!!"
```

```python
# 1. init is a first method/function which automatically gets called when you use name of the class
# for creating an object.

# 2. self keyword is the reference/address of the object, It is automatically passed in the object methods/functions
# like in init it gets passed when you use the name of the class, similarly in bark() it gets passed because
# you used yuki. and that makes python automatically understand that it has to pass the address of yuki.
```

# Inheritance

```python
Inheritance is a concept in Object Oriented Programming which basically deals with how we can inherit
the code from the existing class and write some more features on top of it.
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} Some sound!!"

    
class Cat(Animal):
    def speak(self):
        return f"{self.name} Meow"
    

class Cow(Animal):
    def speak(self):
        return f"{self.name} Moo"
    

class BluAnimal(Animal):
    def enlarge(self):
        return f"Enlarging ..... ........."
```

```python
farms = [Cow('Meghna'), Cat('Persia'), BluAnimal('Blu-exp-234')]
```

```python
for animal in farms:
    print(animal.speak())
```

```python
blu = farms[-1]
```

```python
blu.enlarge()
```

```python
meghna = farms[0]
```

```python
meghna.enlarge()
```

```python
The order of finding the method like speak() in a class inheritance hierarchy is known as
MRO(Method Resolution Order)

a = A(B)

a.method1()

python will try to search for method1 in class A but if not found it will try to go to it's parent B.
If not found even there it will go even further like this in the chain.
```

```python
class Shape:
    def area(self):
        raise NotImplementedError
```

```python
class Square(Shape):
    def __init__(self, name, side):
        self.name = name
        self.side = side
    
    def area(self):
        return self.side * self.side
```

```python
class Rectangle(Shape):
    def __init__(self, name, w, h):
        self.name = name
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h
```

```python
class Circle(Shape):
    def __init__(self, name, r):
        self.name = name
        self.r = r
    
    def area(self):
        return 3.14 * (self.r**2)
```

```python
shape1 = Square('Square Sharma', 5)
```

```python
shape2 = Rectangle("Rectangle Singh", 4, 5)
```

```python
shape3 = Circle('Circle Kumari', 9)
```

```python
for shp in [shape1, shape2, shape3]:
    print(f'Area of shape {shp.name} is {shp.area()}')
```

```python
#Encapsulation

class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
    
    def get_balance(self):
        return self.balance

    def withdraw(self, amount):
        if amount > 0:
            self.balance -= amount
```

```python
class HDFCBankAccount(BankAccount):
    name = 'HDFC'
    
    @classmethod
    def get_name(cls):
        return cls.name
    
    @classmethod
    def set_name(cls, name):
        cls.name = name
```

```python
sbi_account_hitakshi = BankAccount(2000)
```

```python
sbi_account_hitakshi.deposit(200)
```

```python
sbi_account_hitakshi.get_balance()
```

```python
asraful_hdfc_account = HDFCBankAccount(1700)
```

```python
asraful_hdfc_account.withdraw(200)
```

```python
asraful_hdfc_account.get_balance()
```

```python
sbi_account_hitakshi.get_balance()
```

```python
# All the above methods where we are taking self keyword
# and that self keyword is pointing to an object.
# All of these are object methods.

# Object methods are responsible for interacting with the object data and do some
# operations on it.
```
