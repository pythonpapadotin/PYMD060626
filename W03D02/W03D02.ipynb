```python
Object Attribute:- Attribute which changes with every object. Every object has it's own
    balance

Class Attribute:- Attribute which does not change with every object. They are common for all objects.
    name_of_bank
```

```python
Object Methods:- Those methods which are used for doing operations on object attribute.
    They will have a self as first argument.

Class Methods:- Those methods which are used for doing operations on class attributes.
    They will have a cls as first argument
    They will have @classmethod on top of it.
```

```python
Static Methods:- These methods do not do operation on either of them(object and class attributes)
                 They will not have a cls or self as the first argument
                 They will have @staticmethod on top of it.
```

```python
Object Methods also have a special type of methods which are known as dunder methods/magic methods
dunder = double underscore
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance  # object attribute
        
    @staticmethod
    def do_calculation(arg1, arg2, operation):
        if operation == 'add':
            return arg1 + arg2
        elif operation == 'sub':
            return arg1 - arg2
        else:
            raise Excpetion('Error')
    
    def deposit(self, amount):  # object method
        if amount > 0:
            self.balance = BankAccount.do_calculation(self.balance, amount, 'add')
    
    def get_balance(self): # object method
        return self.balance

    def withdraw(self, amount): # object method
        if amount > 0 and self.balance >= amount:
            self.balance = BankAccount.do_calculation(self.balance, amount, 'sub')
        else:
            raise Exception('Balance not enough!!')
```

```python
class BankAccount:
    def __init__(self, first_name, last_name, account_number, ifsc_code, balance):
        self.balance = balance  # object attribute
        self.first_name = first_name # object attribute
        self.last_name = last_name # object attribute
        self.account_number = account_number # object attribute
        self.ifsc_code = ifsc_code # object attribute
        
    def __str__(self):
        return f" Name = {self.first_name} {self.last_name}\n Account Number = {self.account_number}\n IFSC Code = {self.ifsc_code}\n Balance = {self.balance}"

    @staticmethod
    def do_calculation(arg1, arg2, operation):
        if operation == 'add':
            return arg1 + arg2
        elif operation == 'sub':
            return arg1 - arg2
        else:
            raise Excpetion('Error')
    
    def deposit(self, amount):  # object method
        if amount > 0:
            self.balance = BankAccount.do_calculation(self.balance, amount, 'add')
    
    def get_balance(self): # object method
        return self.balance

    def withdraw(self, amount): # object method
        if amount > 0 and self.balance >= amount:
            self.balance = BankAccount.do_calculation(self.balance, amount, 'sub')
        else:
            raise Exception('Balance not enough!!')
```

# File reading and writing

```python
fp = open("first_file.txt", "r")

content = fp.read()  # read the complete file

print(content)

# closing a file is very important
fp.close()
```

```python
Why is closing an file important?
You open a file in memory that is given to the python program by ram.
If you open large files, large portion of this memory will be occupied by those files.
If you do not want memory issues, it's better to close the file.
```

```python
# This is internally defined context manager for closing the file automatically whenever you come out of the
# indentation

with open("first_file.txt", "r") as fp:
    content = fp.read()
    print(content)

print('I am outside indentation. Consider that the file is already closed')
```
